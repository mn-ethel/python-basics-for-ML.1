List, Set, Tuple and Dictionaries

1.List -mutable(changeable)

```python
my_list=[1,2,3,4,5]
print(my_list)
type(my_list)
```

```python
#different data types
my_list = [2,'math',True, 3.5]
print(my_list)
my_list.append(8)
print(my_list)
```

```python
#List indexing
list_1=[1,2,3,5,6,8,2,3]
print(list_1[0])
print(list_1[3])
```

```python
#Length of the list
print(len(list_1))
```

```python
#Initiating an empty list
list_2=[]
print(list_2)
```

```python
list_2.append(5)
print(list_2)
```

```python
#delete an item in a list
list_2=[2,'math',True,3.5,8]
del list_2[3]
print(list_2)
```

```python
#Join two lists
list_3=[1,2,3,4,5]
list_4=[6,7,8,9,10]
list_5=list_3+list_4
print(list_5)
```

2.Tuple- Immutable objects

```python
tuple_1=(5,6,7,8)
print(tuple_1)
type(tuple_1)
```

```python
#Multiple data types
tuple_2=(2,3,4,5,'Machine learning',False)
print(tuple_2)
```

```python
#Converting a list to a tuple
my_list=[3,4,5,6]
print(my_list)
my_tuple=tuple(my_list)
print(my_tuple)
```

```python
#Indexing
print(my_tuple[0])
```

3.Set-mutable

```python
#Curly brackets
my_set={1,2,3,4,5}
print(my_set)
type(my_set)
```

```python
#Convert a list to a set
w=set(list_5)
print(w)
```

4.Dictionary-key-value pair

```python
my_dictionary={'name':'Mercy','age':'21','country':'Kenya'}
print(my_dictionary)
type(my_dictionary)
```

```python
#Calling values using keys
print(my_dictionary['name'])
print(my_dictionary['age'])
print(my_dictionary['country'])
```
