Factorial

```python
number=int(input("Enter the number whose factorial you want:"))
factorial=1
if number ==0:
  print("The factorial of zero is one")
else:
  for i in range(1,number+1):
    factorial=factorial*i
  print("The factorial of " , number, "is" ,factorial)
```

As a function

```python
def factorial_value(num):
  factorial=1
  if num ==0:
    return factorial
  else:
    for i in range(1,num+1):
      factorial=factorial*i
    return factorial
```

```python
print(factorial_value(5)) # Calling the function
```
